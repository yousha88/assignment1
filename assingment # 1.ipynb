```python
#marks of all subjects
computer = int(input("enter your comp marks"))
english = int(input("enter your english marks"))
urdu = int(input("enter your urdu marks"))
physics = int(input("enter your physics marks"))
maths = int(input("enter your maths marks"))
islamiyat = int(input("enter your ist marks"))
#total num of all subjects 
obtain_marks = computer + english + urdu + physics + maths + islamiyat
#finding percentage
percentage = (obtain_marks / 600)*100
print("percentage:", percentage)
if percentage >= 90:
    print("A-1")
elif percentage >= 80:
    print("A")
elif percentage >= 70:
    print("B")
elif percentage >= 60:
    print("C")
elif percentage >= 50:
    print("D")
else:
    print("fail")
```
